# The Force of Impact on a cone Striking a Water Surface (Vertical Entry)

The dimensionless virtual mass k is 
$$
k=\frac{P}{3\rho tan^3(\frac{\theta}{2}) B^{2} U_0^2 (1-\frac{BP}{M_0 U_0^2})}
$$
where
$P$ is the impact force at time t, $\rho$ is the fluid density, $\theta$ is the vertex angle of the cone, $B$ is the depth of penetration, $M_0$ is the mass of the cone, $U_0$ the initial entry velocity
